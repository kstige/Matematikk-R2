## Følger

Naturen rundt oss kan ofte virke som et kaos, men når vi begynner å lete vil vi finne mange skjulte mønster. Slik er det også i matematikken. Matematiske sammenhenger leder ofte til følger av tall som følger etter hverandre i et bestemt system. Dette kaller vi for tallfølger, eller bare følger.

```{admonition} Tallfølger
:class: def
Ei tallfølge (følge) er en liste av tall der tallene avhenger av hverandre av et gitt system. En følge med navn $a$ kan vi skrive som $\{a_n\}$, der $a_n$ er **leddet** med leddnummer $n$ i følgen ($n\in\mathbf{N}$).

$$a_1, a_2, a_3, ..., a_n, ...$$

Vi skiller mellom **endelige** og **uendelige** følger.

**Endelig følge:** $2, 4, 6, 8$

**Uendelig følge:** $2, 4, 6, 8, ...$
```

For å oppgi en følge trenger vi metoder for å oppsummere sammenhengen mellom leddene. Til det kan vi bruke **rekursive** eller **eksplisitte** formler.

### Rekursive formler
Ordet rekursiv betyr noe som gjentas. Mange følger oppgis med det vi kaller en rekursiv formel.

```{admonition} Rekursiv formel
:class: def
En formel er rekursiv dersom formelen for et ledd avhenger av et eller flere tidligere ledd i følgen.

Formelen er ofte slik at $a_{n}$ avhenger av leddet $a_{n-1}$. I tillegg trenger vi et ledd i følgen (gjerne det første) for at formelen skal gi en spesifikk følge.
```

````{admonition} Eksempel 1
:class: eksempel
Tallfølgene $\{a_n\}$ og $\{b_n\}$ er gitt rekursivt ved:

$a_n = a_{n-1} +2, \hspace{20px} n\geq 2, \hspace{20px} a_1 = 2$

$b_{n+1} = 2b_n - 1, \hspace{20px} n \geq 1, \hspace{20px} b_1 = 4$

Finn de fire første leddene i tallfølgene.

```{admonition} Løsning
:class: ekslos, dropdown

Vi ser at for begge følgene, så avhenger et ledd av det forrige leddet i følgen. For å finne leddene må vi derfor beregne et og et ledd.

**a)**

$\begin{align}
a_1 &= 2 \\
a_2 &= a_1 + 2 = 2 + 2 = 4 \\
a_3 &= a_2 + 2 = 4 + 2 = 6 \\
a_4 &= a_3 + 2 = 6 + 2 = 8
\end{align}$

**b)**

$\begin{align}
b_1 &= 4 \\
b_2 &= 2\cdot b_1 - 1 = 2\cdot 4 - 1 = 7 \\
b_3 &= 2\cdot b_2 - 1 = 2\cdot 7 - 1 = 13 \\
b_4 &= 2\cdot b_3 - 1 = 2\cdot 13 - 1 = 25
\end{align}$
```
````

Bruken av rekursive formler er forholdsvis rett fram så lenge man holder styr på at de nedsenkede $n$-ene beskriver nummer på ledd i rekken og ikke regneoperasjoner i formlene. 

Før vi går over til eksplisitte formler tar vi også et eksempel der vi skal finne den rekursive formelen fra ei gitt tallfølge.

````{admonition} Eksempel 3
:class: eksempel
Finn den rekursive formelen for disse tallfølgene

**a)** $20, 16, 12, 8, ...$

**b)** $1, -3, 9, -27, ...$

**c)** $1, 3, 12, 148, ...$

```{admonition} Løsning
:class: ekslos, dropdown
Vi har ingen bestemt metode vi kan bruke for å finne rekursive formler for følger. Vi må selv lete etter system ved å prøve oss fram.

**a)**

$\begin{align}
a_1 &= 20 \\
a_2 &= 16 = 20 - 4 = a_1 - 4 \\
a_3 &= 12 = 16 - 4 = a_2 - 4 \\
a_4 &= 8 = 12 - 4 = a_3 - 4 \\
\end{align}$

Her ser vi at følgen er gitt ved $a_n = a_{n-1} - 4, \hspace{20px} n\geq 2, \hspace{20px} a_1=20$.

**b)**

$\begin{align}
b_1 &= 1 \\
b_2 &= -3 = -3\cdot 1 = -3b_1 \\
b_3 &= 9 = -3\cdot (-3) = -3b_2 \\
b_4 &= -27 = -3\cdot 9 = -3b_3 \\
\end{align}$

Her ser vi at følgen er gitt ved $b_n = -3b_{n-1}, \hspace{20px} n\geq 2, \hspace{20px} b_1=1$.

**c)**

$\begin{align}
c_1 &= 1 \\
c_2 &= 3 = 1^2+2 = a_1^2 + 2 \\
c_3 &= 12 = 3^2+3 = a_2^2 + 3 \\
c_4 &= 148 = 12^2+4 = a_3^2 + 4 \\
\end{align}$

Den her er ikke lett å komme fram til, men vi ser at følgen er gitt ved $c_n = c_{n-1}^2 + n, \hspace{20px} n\geq 2, \hspace{20px} c_1=1$.

```
````

### Eksplisitte formler
Eksplisitte formler er vi mer vant til fra tidligere. Det er formler der vi kan regne ut ledd $n$ uten å kjenne verdiene til de tidligere leddene. Kjente eksempler er formlene for partall, oddetall og kvadrattall:

**Partall:** $p_n = 2\cdot n$

**Oddetall:** $o_n = 2\cdot n - 1$

**Kvadrattall:** $k_n = n^2$

Det finnes ingen bestemt metode for å finne eksplisitte formler for følger. Vi må bruke kreativitet og lete etter system og sammenhenger. Ofte vil det være gunstig å skrive opp de første leddene i følgen og *splitte* de opp i liknende regnestykker.

````{admonition} Eksempel 4
:class: eksempel
Finn den eksplisitte formelen for følgen $a_n = a_{n-1} - 4, \hspace{20px} n\geq 2, \hspace{20px} a_1=20$ som vi så på i eksempel 3.

```{admonition} Løsning
:class: ekslos, dropdown
Vi har ingen bestemt metode vi kan bruke for å finne rekursive formler for følger. Vi må selv lete etter system ved å prøve oss fram.

$\begin{align}
a_1 &= 20 = 20 - 4\cdot 0\\
a_2 &= a_1 - 4 = 20 - 4\cdot 1\\
a_3 &= a_2 - 4 = a_1 - 4 - 4 = 20 - 4\cdot 2\\
a_4 &= a_3 - 4 = a_1 - 4\cdot 2 - 4 = 20 - 4\cdot 3\\
\end{align}$

Her ser vi at alle leddene kan skrives på formen $a_n = 20-4n$. Dette er den eksplisitte formelen for følgen.
```
````

I eksempelet over kom vi fram til en eksplisitt formel ved å lete etter et system. Det er viktig å merke seg at vi ikke har bevist formelen, men kun *anskueliggjort* at det er en riktig formel for følgen. En metode for å bevise at denne formelen stemmer skal vi se på når vi kommer til induksjonsbevis.
